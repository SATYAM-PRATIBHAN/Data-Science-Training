# **Loyal Customers**

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Reads the data
df = pd.read_csv("cust_satisfaction.csv")
df
```

```python
# Provides the information about whole database columns
df.info() 
```

```python
# Provides the statistical Data
df.describe()
```

```python
# Provides all the null values of every column
df.isnull().sum()
```

```python
# Drops every null value
df.dropna(inplace = True)
```

```python
# Prints sum of all null values
df.isnull().sum()
```

```python
# Provides number of all duplicated values
df.duplicated().sum()
```

```python
# Deletes all the duplicate values and provides number of all duplicated values
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

```python
# Provides unique values
print(df['Customer Type'].unique())
print(df['Age'].unique())
print(df['Gender'].unique())
print(df['Type of Travel'].unique())
print(df['Class'].unique())
print(df['satisfaction'].unique())
print(df['Inflight entertainment'].unique())
print(df['Baggage handling'].unique())
print(df['Cleanliness'].unique())
```

```python
# Creates a dictinory of two keys for loyal customers {dt.keys() : Provides keys, dt.values() : Provides values}
dt  = {'Loyal Customer' : 1, 'disloyal Customer' : 0};
df['Customer Type'] = df['Customer Type'].map(dt)
```

```python
# Print unique values of Customer Type
df['Customer Type'].unique()
```

```python
# Creates a dictinory of two keys for Gerders
dtGender  = {'Male' : 1, 'Female' : 0};
df['Gender'] = df['Gender'].map(dtGender)
df['Gender'].unique()
```

```python
# Creates a dictinory of two keys for type of travel
dtTypeOfTravel  = {'Personal Travel' : 1, 'Business travel' : 0};
df['Type of Travel'] = df['Type of Travel'].map(dtTypeOfTravel)
df['Type of Travel'].unique()
df
```

```python
drop_tables = ["Class", "satisfaction", "Baggage handling", "Inflight entertainment", "Cleanliness"]
df.drop(columns = drop_tables, inplace = True)
df
```

```python
x = df.drop('Customer Type', axis = 1)
y = df[['Customer Type']]
```

```python
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
y_test
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train) * 100
```

```python
lr.score(x_test, y_test) * 100
```

```python
pred = lr.predict(x_test)
pred
```

```python
y_test['pred'] = pred.ravel()
y_test
```

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error
print("MSE : ", mean_squared_error(y_test['Customer Type'], y_test['pred']))
print("MAE : ", mean_absolute_error(y_test['Customer Type'], y_test['pred']))
print("RMSE : ", root_mean_squared_error(y_test['Customer Type'], y_test['pred']))
```

```python
y_test['pred'] = (y_test['pred']  > 0.5).astype(int)
y_test['Customer Type'] = (y_test['Customer Type']  > 0.5).astype(int)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, precision_score, accuracy_score, recall_score
print("Confusion Matrix : ",confusion_matrix(y_test['Customer Type'], y_test['pred']))
print("Precision Score : ",precision_score(y_test['Customer Type'], y_test['pred']))
print("F1 Score : ", f1_score(y_test['Customer Type'], y_test['pred']))
print("Accuracy Score : ", accuracy_score(y_test['Customer Type'], y_test['pred']))
print("Recall Score : ", recall_score(y_test['Customer Type'], y_test['pred']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
cm = confusion_matrix(y_test['Customer Type'], y_test['pred'])
cm
```

```python
sns.heatmap(cm, annot=True, cmap="Reds", fmt='d')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
